```python
def gradient_descent(x, y, iterations=20000):
    n = len(x)
    learning_rate = 0.01
    m = 0
    c = 0

    for i in range(iterations):
        y_pred = m * x + c
        slope_def = np.sum((y_pred - y) * x) / n
        slope_int = np.sum(y_pred - y) / n
        cost = (np.sum(y_pred - y)**2)/2*n

        m = m - learning_rate * slope_def
        c = c - learning_rate * slope_int

        print("cost {}, m {}, c {}, iteration {} ".format(cost, m, c, i))

    return m, c
```

```python
i =0;
m ,c =gradient_descent(df['YearsExperience'], df['Salary']);
```

```python
def predict(x_value):
    return m*x_value + c
```
