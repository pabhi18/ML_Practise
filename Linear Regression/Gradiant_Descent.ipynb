```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
```

```python
def compute_cost(x, y, w, b):
    m = x.shape[0]
    cost_sum = 0
    for i in range (m):
        f_wb = w*x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum = cost + cost_sum
        
    total_cost = cost_sum/2*m
    
    return  total_cost
```

```python
def compute_gradient(x, y, w, b):
    m = x.shape[0]
    gradiant_sum = 0
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        f_wb = w*x[i] + b
        dj_dw_i = (f_wb - y[i])*x[i]
        dj_db_i = (f_wb - y[i])
        dj_dw += dj_dw_i
        dj_db += dj_db_i
    dj_dw = dj_dw/m
    dj_db = dj_db/m
    return dj_dw, dj_db
```

```python
def gradiant_descent(x, y, w, b, alpha, iterations):
    dj_dw, dj_db = compute_gradient(x, y, w, b)
    for i in range(iterations):
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        
        cost = compute_cost(x, y, w, b)
        print("w:", w, "b:", b, "Cost:", cost, "iteration:", i)
        
    return w, b
```

```python
alpha = 0.0001
iterations = 2922
w = 0
b = 0
w, b = gradiant_descent(x_train, y_train, w, b, alpha, iterations)
```
