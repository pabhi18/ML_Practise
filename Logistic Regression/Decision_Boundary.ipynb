## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing Dataset

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1) 
```

## Plotting Data

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.axis([0, 4, 0, 3.5])
markers = {0: 'X', 1: 'o'}
plot = sns.scatterplot(x = X[:,0], y = X[:,1], hue = y[:, 0], style=y[:, 0], markers=markers, ax = ax)
plot.set(xlabel ="x0", ylabel = "x1")
```

## Plotting Decision Boundary 

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.axis([0, 4, 0, 3.5])
x0 = np.arange(0,6)
x1 = 3 - x0
ax.plot(x0,x1, c="b")
ax.fill_between(x0,x1, alpha=0.2)
markers = {0: 'X', 1: 'o'}
plot = sns.scatterplot(x = X[:,0], y = X[:,1], hue = y[:, 0], style=y[:, 0], markers=markers, ax = ax)
plot.set(xlabel ="x0", ylabel = "x1")
```
